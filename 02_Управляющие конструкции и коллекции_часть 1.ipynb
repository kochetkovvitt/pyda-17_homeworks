##### Задание 1
Дана переменная, в которой хранится слово из латинских букв. Напишите код, который выводит на экран:
- среднюю букву, если число букв в слове нечетное;
- две средних буквы, если число букв четное.

```python
word = 'testtest'
n = len(word)

if n % 2 == 0:
    print(word[int(n/2)-1:int(n/2)+1])
else:
    print(word[n//2])
```

##### Задание 2
Напишите программу, которая последовательно запрашивает у пользователя числа (по одному за раз)  
и после первого нуля выводит сумму всех ранее введенных чисел.

```python
input_list = []
while True:
    num = int(input('Введите число: '))
    if num == 0:
        print(sum(input_list))
        break
    else:
        input_list.append(num)
```

##### Задание 3
Мы делаем MVP dating-сервиса, и у нас есть список парней и девушек.  
Выдвигаем гипотезу: лучшие рекомендации мы получим,если просто отсортируем имена по алфавиту  
и познакомим людей с одинаковыми индексами после сортировки! Но мы не будем никого знакомить,  
если кто-то может остаться без пары:

```python
boys = ['Peter', 'Alex', 'John', 'Arthur', 'Richard'] 
girls = ['Kate', 'Liza', 'Kira', 'Emma', 'Trisha']

if len(boys) != len(girls):
    print('Внимание, кто-то может остаться без пары!')
else:
    print('Идеальные пары:')
    for i in zip(sorted(boys), sorted(girls)):
        print(i[0],'и', i[1])
```

##### Задание 4
У нас есть список, содержащий информацию о среднедневной температуре в Фаренгейтах за произвольный период по странам.  
Необходимо написать код, который рассчитает среднюю температуру за период в Цельсиях(!) для каждой страны.



```python
countries_temperature = [
['Thailand', [75.2, 77, 78.8, 73.4, 68, 75.2, 77]],
['Germany', [57.2, 55.4, 59, 59, 53.6]],
['Russia', [35.6, 37.4, 39.2, 41, 42.8, 39.2, 35.6]],
['Poland', [50, 50, 53.6, 57.2, 55.4, 55.4]]
]

print('Средняя температура в странах:')
for i in countries_temperature:
    c = (sum(i[1]) / len(i[1]) - 32) * 5 / 9
    print(f'{i[0]} - {c:.1f} C')
```

##### Задание 5
Дан поток логов по количеству просмотренных страниц для каждого пользователя.  
Вам необходимо написать алгоритм, который считает среднее значение просмотров на пользователя.  
Т. е. надо посчитать отношение суммы всех просмотров к количеству уникальных пользователей.

```python
stream = [
    '2018-01-01,user1,3',
    '2018-01-07,user1,4',
    '2018-03-29,user1,1',
    '2018-04-04,user1,13',
    '2018-01-05,user2,7',
    '2018-06-14,user3,4',
    '2018-07-02,user3,10',
    '2018-03-21,user4,19',
    '2018-03-22,user4,4',
    '2018-04-22,user4,8',
    '2018-05-03,user4,9',
    '2018-05-11,user4,11']

users = []
views = []

for i in stream:
    log = i.split(',')
    if log[1] not in users:
        users.append(log[1])
    views.append(int(log[2]))
    
print(f'Среднее количество просмотров на уникального пользователя: {sum(views)/len(users):.2f}')
```

##### Задание 6
Необходимо у пользователя запрашивать набор чисел разделенных пробелом.  
В результате должны выводиться числа, которые повторяются в вводе более одного раза.

```python
num = input('Введите числа: ').split(' ')
repeated = []

for i in range(len(num)):
    if num[i] in num[i+1:] and num[i] not in repeated:
        repeated.append(num[i])

print(' '.join(repeated))
```
