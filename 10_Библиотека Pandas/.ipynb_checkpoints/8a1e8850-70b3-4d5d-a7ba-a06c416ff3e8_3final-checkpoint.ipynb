–ó–¥–µ—Å—å –º–æ–∂–Ω–æ –ø—Ä–æ—Å—Ç–∞–≤–∏—Ç—å –≥–æ–¥, –∞ –ø—Ä–æ–ø—É—Å–∫–∏ –ø–æ –æ—Ü–µ–Ω–∫–∞–º –∑–∞–º–µ–Ω–∏—Ç—å –Ω–∞–ø—Ä–∏–º–µ—Ä –Ω–∞ –º–µ–¥–∏–∞–Ω–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –∏–∑ –∏–º–µ—é—â–∏—Ö—Å—è.<br>
–ê–≤—Ç–æ–º–∞—Ç–∏–∑–∏—Ä—É–µ–º –ø—Ä–æ—Ü–µ—Å—Å:
</div>

```python
for name_ in tqdm(df.name.unique()):
    data = df[df.name == name_]
    
    if len(data) > 1 and data.isnull().values.any():
    # –µ—Å–ª–∏ –≤ –≤—ã–±–æ—Ä–∫–µ –±–æ–ª—å—à–µ –æ–¥–Ω–æ–π —Å—Ç—Ä–æ–∫–∏ –∏ –µ—Å—Ç—å –ø—Ä–æ–ø—É—Å–∫–∏
        data.year_of_release.fillna(data.year_of_release.max(), inplace=True)
        data.critic_score.fillna(data.critic_score.median(), inplace=True)
        data.user_score.fillna(data.user_score.median(), inplace=True)    
        rating_ = data.rating.fillna('-').max()
        if rating_ != '-':
            data.rating.fillna(rating_, inplace=True)
        df[df.name == name_] = data
# Wall time: 00:59
```

<div class="alert alert-block alert-info">
–£–∂–µ –Ω–µ–º–Ω–æ–≥–æ –ª—É—á—à–µ.<br>
–¢–µ–ø–µ—Ä—å –ø–æ—Å–º–æ—Ç—Ä–∏–º –Ω–∞ –æ—Å—Ç–∞–≤—à–∏–µ—Å—è –ø—Ä–æ–ø—É—Å–∫–∏ –≤ —Ä–∞–∑—Ä–µ–∑–µ –ø–ª–∞—Ç—Ñ–æ—Ä–º.<br>
–°–∫–æ–ª—å–∫–æ —Ç–∞–º –≤—Å–µ–≥–æ –∏–≥—Ä –∏ –∫–∞–∫–∞—è –¥–æ–ª—è –∏–∑ –Ω–∏—Ö –ø–æ–∫—Ä—ã—Ç–∞ –æ—Ü–µ–Ω–∫–∞–º–∏:
</div>

```python
def show_filled_part_by_platforms(data, subset):
    data_f = data.dropna(subset=subset)
    pt1 = pd.pivot_table(data, 
                   index="platform",
                   values="name",
                   columns="year_of_release",
                   fill_value=0,
                   aggfunc=[len]).sort_values(by=('platform'))
    pt1.columns = [col[1] for col in pt1.columns.values]
    pt2 = pd.pivot_table(data_f, 
               index="platform",
               values="name",
               columns="year_of_release",
               fill_value=0,
               aggfunc=[len]).sort_values(by=('platform'))
    pt2.columns = [col[1] for col in pt2.columns.values]
    display_side_by_side(pt1, pt2 / pt1)
```

```python
show_filled_part_by_platforms(df.query('year_of_release > 2011'), ['critic_score', 'user_score'])
```

–ü–æ—Å–º–æ—Ç—Ä–∏–º, –∫–∞–∫ –≤–ª–∏—è—é—Ç –Ω–∞ –ø—Ä–æ–¥–∞–∂–∏ –≤–Ω—É—Ç—Ä–∏ –æ–¥–Ω–æ–π –ø–æ–ø—É–ª—è—Ä–Ω–æ–π –ø–ª–∞—Ç—Ñ–æ—Ä–º—ã –æ—Ç–∑—ã–≤—ã –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –∏ –∫—Ä–∏—Ç–∏–∫–æ–≤.

<div class="alert alert-block alert-info"> 
–°—Ä–∞–∑—É –∞–≤—Ç–æ–º–∞—Ç–∏–∑–∏—Ä—É–µ–º, –æ–±–µ—Ä–Ω—É–≤ –ø—Ä–æ—Ü–µ—Å—Å –≤ —Ñ—É–Ω–∫—Ü–∏—é:<br> 
</div>

```python
def show_score_corr(platform):
    data = df.query('platform == @platform')
    
    print(f'\n{platform}. –ö–æ—Ä—Ä–µ–ª—è—Ü–∏—è –º–µ–∂–¥—É critic_score –∏ all_sales: ', end='')
    print(data['critic_score'].corr(data['all_sales']))
    
    print(f'\n{platform}. –ö–æ—Ä—Ä–µ–ª—è—Ü–∏—è –º–µ–∂–¥—É user_score –∏ all_sales: ', end='')
    print(data['user_score'].corr(data['all_sales']))
    
    fig, ax = plt.subplots(1, 2, figsize=(11, 4))
    sns.stripplot(x='critic_score', y='all_sales', data=data, ax=ax[0])
    sns.stripplot(x='user_score', y='all_sales', data=data, ax=ax[1])
    fig.show()
```

```python
show_score_corr('XOne')
```

–ö–æ—Ä—Ä–µ–ª—è—Ü–∏—é –º–µ–∂–¥—É –æ—Ç–∑—ã–≤–∞–º–∏ –∫—Ä–∏—Ç–∏–∫–æ–≤ –∏ –ø—Ä–æ–¥–∞–∂–∞–º–∏ 0.42 –º–æ–∂–Ω–æ –æ—Ü–µ–Ω–∏—Ç—å –∫–∞–∫ —Å—Ä–µ–¥–Ω—é—é –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω—É—é, –∞ –≤–æ—Ç –Ω–∞ –æ—Ü–µ–Ω–∫–∏ –∏–≥—Ä–æ–∫–æ–≤ –≤–∏–¥–∏–º–æ –±–æ–ª—å—à–µ –≤–ª–∏—è—é—Ç –∏—Ö –æ–∂–∏–¥–∞–Ω–∏—è –∏ —ç–º–æ—Ü–∏–∏ - –∫–æ—Ä—Ä–µ–ª—è—Ü–∏—è –Ω–∏–∑–∫–∞—è –∏ –∫ —Ç–æ–º—É –∂–µ –æ—Ç—Ä–∏—Ü–∞—Ç–µ–ª—å–Ω–∞—è -0.07. 

–î–ª—è —Å—Ä–∞–≤–Ω–µ–Ω–∏—è –ø–æ—Å–º–æ—Ç—Ä–∏–º –Ω–∞ –∞–Ω–∞–ª–æ–≥–∏—á–Ω—ã–µ –ø–æ–∫–∞–∑–∞—Ç–µ–ª–∏ –Ω–∞ –¥—Ä—É–≥–æ–π –ø–ª–∞—Ç—Ñ–æ—Ä–º–µ - 3DS:
